### 3.2 hash function
hashhash table

```python
# 
phone_book = {}

# 
def hash_function(name):
    # ASCII
    hash_value = sum(ord(char) for char in name) % 10
    return hash_value

# 
def add_contact(name, phone_number):
    slot = hash_function(name)
    phone_book[slot] = phone_number

# 
def find_contact(name):
    slot = hash_function(name)
    if slot in phone_book:
        return phone_book[slot]
    else:
        return "Contact not found"

# 
add_contact("Alice", "123-456-7890")
add_contact("Bob", "987-654-3210")
add_contact("Charlie", "555-123-4567")

# 
print(find_contact("Alice"))  # 123-456-7890
print(find_contact("Charlie"))  # 555-123-4567
print(find_contact("Dave"))  # Contact not found "Dave" 
```

## 2. Selection Sort
Selection Sort
- 
- 
- 



\
\
 O(n^2) n 

O(N^2)\
Swaps N-1\
Comparisons (N-1) + (N-2) + (N-3) + ... + 1(N * (N-1)) / 2\
Assignments 1(N * (N-1)) / 2 + 1

\
   +  + \
  (N-1) + [(N * (N-1)) / 2] + [(N * (N-1)) / 2 + 1]

\
  (N * (N-1)) + [(N * (N-1)) / 2] + 1
NN^2**N^2**

```python
def selection_sort(arr):
    # 
    for i in range(len(arr)):
        # 
        min_index = i

        #   (N-1)+(N-2)+...+1-->N^2
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[min_index]:
                min_index = j

        #   N-1
        arr[i], arr[min_index] = arr[min_index], arr[i]

# 
numbers = [64, 25, 12, 22, 11]
selection_sort(numbers)
print(":", numbers)
```

## 3. Bubble Sort
Bubble Sort



-

ON^2,N

`for j in range(0, n - i - 1)`:  n - i - 1  n - i - 1  i \


```python
list_num=[64, 34, 25, 12, 22, 11, 90]

def bubble_sort(my_list):
    
    for i in range(len(my_list)):
        
        for j in range(len(my_list)-i-1):
            
            if my_list[j]>my_list[j+1]:
                my_list[j],my_list[j+1]=my_list[j+1],my_list[j]
    return my_list

bubble_sort(list_num)
```

## 4. Insertion Sort
Insertion Sort

- 
- 
- 

ON^2

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):  # 
        current_element = arr[i]  # 
        j = i - 1  # 

        while j >= 0 and current_element < arr[j]:
            # current elementwhile
            arr[j + 1] = arr[j] # 
            j -= 1

        # 
        arr[j + 1] = current_element

# 
numbers = [64, 34, 25, 12, 22, 11, 90]
insertion_sort(numbers)
print(":", numbers)
```

## 4. Merge Sort
Merge Sort****


- Divide
- Conquer
- Merge

O(nlogn)

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        # 
        merge_sort(left_half)
        merge_sort(right_half)

        i = j = k = 0

        # 
        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1

# 
numbers = [38, 27, 43, 3, 9, 82, 10]
merge_sort(numbers)
print(":", numbers)
```

## 5. Quick Sort
Quick Sort********

:O(nlogn)

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr

    pivot = arr[0]  # 
    left = [x for x in arr[1:] if x <= pivot] #pivotpivot
    right = [x for x in arr[1:] if x > pivot] #pivotpivot

    # 
    return quick_sort(left) + [pivot] + quick_sort(right)

# 
numbers = [38, 27, 43, 3, 9, 82, 10]
sorted_numbers = quick_sort(numbers)
print(":", sorted_numbers)
```
