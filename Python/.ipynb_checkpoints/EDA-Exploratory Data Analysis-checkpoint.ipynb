## 2. Data Preparation and Cleaning
### 2.1 Understanding the data
While the survey responses contain a wealth of information, we'll limit our analysis to the following areas:

- Demographics of the survey respondents and the global programming community 即调查受访者的人口统计信息和全球编程社区
- Distribution of programming skills, experience, and preferences 即编程技能、经验和偏好的统计分布
- Employment-related information, preferences, and opinions 即与就业相关的信息、偏好和观点


```python
# 选出需要分析的列
selected_columns = [
    # Demographics
    'Country',
    'Age',
    'Gender',
    'EdLevel',
    'UndergradMajor',
    
    # Programming experience
    'Hobbyist',
    'Age1stCode',
    'YearsCode',
    'YearsCodePro',
    'LanguageWorkedWith',
    'LanguageDesireNextYear',
    'NEWLearn',
    'NEWStuck',
    
    # Employment
    'Employment',
    'DevType',
    'WorkWeekHrs',
    'JobSat',
    'JobFactors',
    'NEWOvertime',
    'NEWEdImpt'
]

len(selected_columns)   # 20
```

```python
# 将需要分析的列单独copy出dataframe，保证原始数据不受影响
response_df = survey_response[selected_columns].copy()

#由于survey_questions的index是'Column'，所以需要用.loc[]来选取行名为selected_columns的行
questions_df = survey_questions.loc[selected_columns].copy() 

response_df.shape, questions_df.shape
```

### 2.2 Cleaning the data
通过info()可以看到，response_df中有64461行，20列，其中有大部分列都有缺失值，后面需要对这些缺失值进行处理。

只有两列被检测为numeric列（Age，WorkWeekHrs）。为了使我们的分析更容易，让我们将其他一些列转换为数值数据类型，同时忽略任何非数字值。非数字将被转换为NaN。

检查response_df,可以看到，Age1stCode，YearsCode，YearsCodePro这3列的dtype都是object，但是值都是数字，所以需要将其转换为数值类型。
- `to_numeric()`: 将其转换为数值类型，但是会报错，因为这些列中存在一些非数字的值，所以需要加上`errors='coerce'`，将非数字的值转换为NaN。


```python
response_df.info()
```

```python
# 将Age1stCode，YearsCode，YearsCodePro这3列转换为numeric类型

response_df['Age1stCode'] = pd.to_numeric(response_df['Age1stCode'], errors='coerce')
response_df['YearsCode'] = pd.to_numeric(response_df['YearsCode'], errors='coerce')
response_df['YearsCodePro'] = pd.to_numeric(response_df['YearsCodePro'], errors='coerce')
```

```python
# 查看此时的response_df的数据情况
response_df.describe() #decribe()只能看到数值类型的列的统计信息
```

可以看到经过数值转化，describe()统计了numeric类型列的信息，其中存在一些问题：
- `Age`列的最小值为1，最大值为279，这是不合理的，因为这是调查程序员的年龄，应该删去 <10 和 >100 的数据
- `WorkWeekHrs`列的最小值为1，最大值为475，这也是不合理的，应该删去 >140 的数据

```python
# 删除Age列中 <10 和 >100 的数据
response_df.drop(response_df[response_df.Age < 10].index, inplace=True)
response_df.drop(response_df[response_df.Age > 100].index, inplace=True)

# 删除WorkWeekHrs列中 >140 的数据
response_df.drop(response_df[response_df.WorkWeekHrs > 140].index, inplace=True)

# response_df.describe()
```

接下来关于`Gender`列，我们可以看到，Gender列中存在多个性别，我们需要将其转换为`Man`、`Woman` 和 
`Non-binary, genderqueer, or gender non-conforming`三种性别。

```python
response_df.Gender.value_counts()
```

```python
response_df.where(~(response_df.Gender.str.contains(';', na=False)), other=np.nan, inplace=True)
# where(condition, other=nan, inplace=False, axis=None, level=None, errors='raise', try_cast=False)
# 表示如果condition为真，则返回原来的值，否则返回other的值
# ~表示取反,即如果Gender列中不包含';'，则返回原来的值，否则返回np.nan
# 即将多选项直接转换为NaN，仅留下单选一种性别的行
```

```python
response_df.Gender.value_counts()
```

可以看到仍然存在很多NaN值，需要进一步处理
对于numeric类型列：可以用平均值填充，`.fillna()`方法
其他非数值型列可以选择删除`dropna()`方法

#### 分析EdLevel与Gender的关系
如果要查看不同性别的学历情况，按 'Gender' 对 'EdLevel' 进行分组，你需要重构数据以便每行包含一个 'Gender' 和一个 'EdLevel' 值。使用 groupby 和 value_counts 是一种实现这一点的方法。然后，你可以使用 hue 参数在 barplot 中按 'Gender' 分组

```python
# # grouped_response_df = response_df.groupby('Gender')['EdLevel'].value_counts(normalize=True).rename('percentage').reset_index()
# 
# d1=response_df.groupby('Gender')['EdLevel'] # 得到一个groupby对象
# # print(type(d1)) #<class 'pandas.core.groupby.generic.SeriesGroupBy'>
# 
# d1.value_counts(normalize=True) # 此时groupby对象变为series
# d1.value_counts(normalize=True).rename('percentage')  # 将value_counts()方法返回的series重命名为percentage
# d1.value_counts(normalize=True).rename('percentage').reset_index() # reset_index()将series转换为dataframe
```

从图中可以看到，最多人获取了Bachelor's degree，其次是Master's degree，最少的是Primary/elementary school。
而从性别角度，男性和女性的学历分布基本相同，女性在Bachelor和Master的占比均比男性高。

2. 添加新列`earned_degree`，将EdLevel列中的值进行分组，分为Yes和No两类。
- Yes：表示获得degree
- No：表示未获得degree

使用`earned_degree`列来分组，然后查看NEWEdImpt列的分布情况

### 3.6 Correlation between Age and YearsCodePro

使用散点图Scatter Plot来查看Age和YearsCodePro的关系

```python
sns.scatterplot(x='Age', y='YearsCodePro',hue='Hobbyist',  data=response_df)
plt.xlabel("Age")
plt.ylabel("Years of professional coding");
```

#### 分析Age和YearsCodePro的相关性
首先计算相关系数，使用`corr()`方法，然后使用`heatmap()`方法可视化相关系数

```python
# 使用pearson相关系数
from scipy.stats import pearsonr

s_df= response_df.dropna()
data1 = s_df['Age']
data2 = s_df['YearsCodePro']
# calculate Pearson's correlation
corr,_ = pearsonr(data1, data2)
print('Pearsons correlation: %.3f' % corr)

# 使用pandas中的corr()方法
corr_matrix=response_df[['Age', 'YearsCodePro']].corr()
corr_matrix
```

```python
# 使用heatmap可视化
sns.heatmap(corr_matrix, annot=True, cmap='Greens');
```
